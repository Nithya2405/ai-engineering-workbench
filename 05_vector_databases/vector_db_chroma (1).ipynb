# 1. Initialize the "Architect" - A Persistent Local DB
This creates a folder named 'my_local_db' to store your data

```python
client = chromadb.PersistentClient(path="./my_local_db")
```

# 2. Use a default embedding function (Local & Free)

```python
default_ef = embedding_functions.DefaultEmbeddingFunction()
```

# 3. Create a 'Collection' (Think of this as a Table in SQL)
We use get_or_create so it doesn't error if you run it twice

```python
collection = client.get_or_create_collection(
    name="college_knowledge",
    embedding_function=default_ef
)
```

# 4. Our "Document Chunks" (The memory we want to give the AI)

```python
documents = [
    "The library is located next to the student center.",
    "Final exams start on May 15th this semester.",
    "The AI Lab has 24/7 access for engineering students.",
    "The cafeteria serves pizza every Friday."
]
ids = ["doc1", "doc2", "doc3", "doc4"]
metadatas = [{"type": "map"}, {"type": "calendar"}, {"type": "access"}, {"type": "food"}]
```

# 5. Add to the Vector DB

```python
collection.add(
    documents=documents,
    ids=ids,
    metadatas=metadatas
)

print(f"✅ Successfully stored {collection.count()} chunks in ChromaDB!")
```

# 6. Test a quick query

```python
results = collection.query(
    query_texts=["When are the finals?"],
    n_results=1
)

print(f"\n🔍 Query Result: {results['documents'][0][0]}")
```

```python
# Day 5: Vector DB Initialization (ChromaDB)
# Concept: Building "Long-term Memory" for AI

import chromadb
from chromadb.utils import embedding_functions

# 1. ARCHITECT STEP: Initialize persistent storage
# This saves'brain' to a folder so it doesn't disappear
client = chromadb.PersistentClient(path="./chroma_db")

# 2. Define the Embedding Function (Free & Local)
# This converts your text into 384-dimensional vectors
emb_fn = embedding_functions.SentenceTransformerEmbeddingFunction(model_name="all-MiniLM-L6-v2")

# 3. Create the Collection (The "Library")
collection = client.get_or_create_collection(name="college_docs", embedding_function=emb_fn)

# 4. Store Chunks (The "Knowledge")
collection.add(
    documents=[
        "The AI lab is on the 3rd floor, Room 302.",
        "Professor Smith's office hours are Monday 2PM-4PM.",
        "The scholarship deadline is October 15th."
    ],
    ids=["id1", "id2", "id3"]
)

print(f"✅ Collection count: {collection.count()} documents stored.")

# 5. TEST QUERY: Semantic Retrieval
query = "Where can I find the AI lab?"
results = collection.query(query_texts=[query], n_results=1)

print(f"\n🔍 Query: {query}")
print(f"🤖 Retrieved Memory: {results['documents'][0][0]}")
```
