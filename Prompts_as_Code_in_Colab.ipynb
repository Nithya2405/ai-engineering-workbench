For Day 16, running this in Colab requires a slightly different approach than a local environment because Colab’s file system is temporary. Since we are moving toward **Prompts as Code**, we need to programmatically build that directory structure so your code has something to read.

Here is the step-by-step implementation for your **02_prompt_management.ipynb**.

---

### 🛠️ Step 1: Initialize the Prompt Directory

Since Colab resets its storage every session, we use Python to create the `prompts/` folder and the YAML files automatically.

```
import os
import yaml

# 1. Create the prompts directory
os.makedirs('prompts', exist_ok=True)

# 2. Define your Prompt versions
prompts = {
    "researcher_v1.yaml": {
        "version": "1.0",
        "role": "Professional Research Assistant",
        "instructions": "You are an expert market analyst. Focus on identifying risks and opportunities. Always provide citations.",
        "output_format": "Markdown"
    },
    "researcher_v2.yaml": {
        "version": "2.0",
        "role": "Critical Risk Auditor",
        "instructions": "You are a pessimistic financial auditor. Find the flaws in every argument. Be concise and brutal.",
        "output_format": "Bullet Points"
    }
}

# 3. Save these to the /prompts folder
for filename, content in prompts.items():
    with open(f"prompts/{filename}", 'w') as file:
        yaml.dump(content, file, default_flow_style=False)

print("✅ Prompt Registry Initialized: prompts/researcher_v1.yaml and v2.yaml created.")
```

### 🧠 Step 2: The Production-Ready Loader

This function is what you’ll use in your agentic logic. It decouples the *what* (the instructions) from the *how* (the Python code).

```
def load_prompt(version):
    file_path = f"prompts/researcher_{version}.yaml"
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"Prompt version {version} not found!")

    with open(file_path, 'r') as file:
        return yaml.safe_load(file)

# Usage: Switching from v1 to v2 is now a single string change
config = load_prompt("v1")

system_message = f"""
Role: {config['role']}
Instructions: {config['instructions']}
Format: {config['output_format']}
"""

print(f"📡 Loaded Version: {config['version']}")
print(f"📝 Active System Prompt:\n{system_message}")
```
