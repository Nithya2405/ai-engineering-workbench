For an LLM to use a tool, the tool must be a standard Python function. We will use **Tavily**, as it is purpose-built for AI Agents—it filters out the "noise" (ads, formatting) and returns only the clean content the LLM needs.

**To build in Colab:**

1. **Secret Key:** Add a new secret in Colab named `TAVILY_API_KEY`. (Get one for free at [tavily.com](https://tavily.com)).
2. **The Tool:** Define a function that takes a query, calls the API, and returns a formatted string of results.

**The Code Logic:**

```
!pip install tavily-python
from google.colab import userdata
from tavily import TavilyClient

# 1. Setup the Search Client
TAVILY_API_KEY = userdata.get('TAVILY_API_KEY')
tavily = TavilyClient(api_key=TAVILY_API_KEY)

def web_search_tool(query: str):
    """
    Searches the live web for the given query and returns a clean summary.
    This is the 'eyes' for our AI Agent.
    """
    print(f"DEBUG: Agent is searching for: {query}")

    # search_depth="advanced" gets deeper data; "basic" is faster.
    search_result = tavily.search(query=query, search_depth="basic", max_results=3)

    # Format the results into a single string for the LLM
    context = ""
    for result in search_result['results']:
        context += f"\nSource: {result['url']}\nContent: {result['content']}\n"

    return context

# Test the tool independently
print(web_search_tool("What is the current stock price of NVIDIA?"))
```
