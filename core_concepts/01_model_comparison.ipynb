```python
!pip install -q -U google-genai groq

import time
from google import genai
from groq import Groq
from google.colab import userdata

# Initialize Clients
gemini_client = genai.Client(api_key=userdata.get('GEMINI_API_KEY'))
groq_client = Groq(api_key=userdata.get('GROQ_API_KEY'))

PROMPT = "Compare 'Inference Latency' vs 'Throughput' in two concise sentences."

def test_gemini():
    start = time.time()
    response = gemini_client.models.generate_content(
        model="gemini-3-flash-preview",
        contents=PROMPT
    )
    return response.text, time.time() - start

def test_groq():
    start = time.time()
    # Using Llama 3.3 70b on Groq for high performance
    response = groq_client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[{"role": "user", "content": PROMPT}]
    )
    return response.choices[0].message.content, time.time() - start

# --- The Comparison Showdown ---
print("STARTING MODEL SHOWDOWN\n")

try:
    g_text, g_time = test_gemini()
    print(f"Gemini 3 Flash: {g_time:.2f}s\n{g_text}\n")

    q_text, q_time = test_groq()
    print(f"Groq (Llama 3.3): {q_time:.2f}s\n{q_text}\n")

    winner = "Groq" if q_time < g_time else "Gemini"
    print(f"🏆 Speed Winner: {winner} (Difference: {abs(g_time - q_time):.2f}s)")

except Exception as e:
    print(f"Error during comparison: {e}")
```
