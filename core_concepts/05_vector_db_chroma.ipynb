# Day 5: Vector DB Initialization
# Logic: Creating a persistent home for our embeddings

```python
!pip install -q chromadb

import chromadb
from chromadb.utils import embedding_functions
```

# 1. Initialize the "Architect" - A Persistent Local DB
This creates a folder named 'my_local_db' to store your data

```python
client = chromadb.PersistentClient(path="./my_local_db")
```

# 2. Use a default embedding function (Local & Free)

```python
default_ef = embedding_functions.DefaultEmbeddingFunction()
```

# 3. Create a 'Collection' (Think of this as a Table in SQL)
We use get_or_create so it doesn't error if you run it twice

```python
collection = client.get_or_create_collection(
    name="college_knowledge",
    embedding_function=default_ef
)
```

# 4. Our "Document Chunks" (The memory we want to give the AI)

```python
documents = [
    "The library is located next to the student center.",
    "Final exams start on May 15th this semester.",
    "The AI Lab has 24/7 access for engineering students.",
    "The cafeteria serves pizza every Friday."
]
ids = ["doc1", "doc2", "doc3", "doc4"]
metadatas = [{"type": "map"}, {"type": "calendar"}, {"type": "access"}, {"type": "food"}]
```

# 5. Add to the Vector DB

```python
collection.add(
    documents=documents,
    ids=ids,
    metadatas=metadatas
)

print(f"✅ Successfully stored {collection.count()} chunks in ChromaDB!")
```

# 6. Test a quick query

```python
results = collection.query(
    query_texts=["When are the finals?"],
    n_results=1
)

print(f"\n🔍 Query Result: {results['documents'][0][0]}")
```
