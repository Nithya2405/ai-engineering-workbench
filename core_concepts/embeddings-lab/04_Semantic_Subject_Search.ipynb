```python
# !pip install sentence-transformers
from sentence_transformers import SentenceTransformer, util

# 1. Load an Open-Source embedding model (SOTA performance) [cite: 79, 80, 81]
model = SentenceTransformer('all-MiniLM-L6-v2')

# 2. Your college subjects
subjects = [
    "Machine Learning",
    "Artificial Intelligence",
    "Data Structures and Algorithms",
    "Operating Systems",
    "Database Management Systems",
    "Deep Learning",
    "Probability and Statistics"
]

# 3. Convert text to vectors (Embeddings) [cite: 97]
embeddings = model.encode(subjects)

# 4. Search for a 'meaning'
query = "I want to learn about neural networks"
query_embedding = model.encode(query)

# 5. Compute similarity
hits = util.semantic_search(query_embedding, embeddings, top_k=3)

print(f"Query: {query}\n")
print("Top 3 Semantically Related Subjects:")
for hit in hits[0]:
    print(f"- {subjects[hit['corpus_id']]} (Score: {hit['score']:.4f})")
```
