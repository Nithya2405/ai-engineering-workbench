```python
!pip install tiktoken

import tiktoken

def calculate_tokens_and_cost(text, model="gpt-4o"):
    # Load the specific encoding for the model
    encoding = tiktoken.encoding_for_model(model)

    # Encode the text into tokens
    tokens = encoding.encode(text)
    token_count = len(tokens)

    # Estimate cost (Using $2.50 per 1M input tokens as an example for GPT-4o)
    cost_per_token = 2.50 / 1_000_000
    estimated_cost = token_count * cost_per_token

    print(f"Model: {model}")
    print(f"Token Count: {token_count}")
    print(f"Estimated Input Cost: ${estimated_cost:.6f}")

# Test it
sample_prompt = "Explain neural networks to me like I'm a five-year-old."
calculate_tokens_and_cost(sample_prompt)
```
