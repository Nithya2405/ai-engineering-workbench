### 🛠️ The Project: "The Token Analytics Utility"

This script does three things:

1. **Counts Tokens:** Uses `tiktoken` (the library used by OpenAI) to show how a model "sees" your text.
2. **Calculates Cost:** Automatically estimates the price based on current API rates.
3. **Visualizes Density:** Shows which parts of your text are "token-heavy."

### Step 2: The Analysis Script

```
import tiktoken
import matplotlib.pyplot as plt

def analyze_tokens(text, model="gpt-4o"):
    # 1. Initialize encoding for the specific model
    encoding = tiktoken.encoding_for_model(model)

    # 2. Encode the text into tokens
    tokens = encoding.encode(text)
    num_tokens = len(tokens)

    # 3. Calculate cost (Example: $2.50 per 1M tokens for GPT-4o input)
    cost_per_1k = 0.0025
    estimated_cost = (num_tokens / 1000) * cost_per_1k

    # 4. Show the results
    print(f"--- Token Report for {model} ---")
    print(f"Text length: {len(text)} characters")
    print(f"Token count: {num_tokens}")
    print(f"Estimated Input Cost: ${estimated_cost:.6f}")

    # 5. Visualizing Tokenization
    # We'll show how the text is split (approximate visualization)
    decoded_tokens = [encoding.decode([t]) for t in tokens]

    print("\nHow the model sees your text (Token Segments):")
    print("|".join(decoded_tokens))

    # 6. Simple Bar Chart
    plt.figure(figsize=(8, 4))
    plt.bar(['Characters', 'Tokens'], [len(text), num_tokens], color=['skyblue', 'salmon'])
    plt.title('Character vs. Token Count')
    plt.ylabel('Count')
    plt.show()

# Test it with a technical sentence
sample_text = "Artificial Intelligence and Machine Learning Engineering is the future of software."
analyze_tokens(sample_text)
```
